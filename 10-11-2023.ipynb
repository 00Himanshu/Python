Write a Python function that takes a list of numbers as input and returns the average of all the even numbers in the list.

```python
def main():
  n=eval(input("Enter the list of numbers: "))
  print(even_avg(n))

def even_avg(x):
  sum=0
  count=0
  for i in x:
    if i%2==0:
      sum+=i
      count+=1
  return sum/count

main()
```

Write a Python Program to Find Factorial of Number Using Recursion.

```python
def main():
  num=int(input("Enter the Number to find Factorial: "))
  print(factorial(num))

def factorial(n):
  f=1
  for i in range(1, n+1):
    f*=i
  return f
main()
```

Create a Python function that calculates the Fibonacci sequence up to a specified limit. The function should return a list of Fibonacci numbers.

```python
def main():
  n=int(input("Enter the number: "))
  print(Fibonacci(n))

def Fibonacci(x):
  lis=[]
  a,b=0,1
  for _ in range (x+1):
    lis.append(a)
    a,b=b, a+b
  return lis

main()
```

Implement a Python function to display the reverse of the number you input.

```python
def main():
  num1 = int(input("Enter first Number: "))
  num2= int(input("Enter Second Number: "))
  print(reverse(num1, num2))

def reverse(n, m):
  a=[]
  for i in range(n,m+1):
    a.append(i)
  return a[::-1]

main()
```

Write a Python function that generates a random password of a specified length. The password should include a combination of uppercase letters, lowercase letters, digits, and special characters.

```python
import random
import string

def main():
    length = int(input("Enter the desired length of the password: "))
    print(password(length))

def password(length=12):
    uppercase_letters = string.ascii_uppercase
    lowercase_letters = string.ascii_lowercase
    digits = string.digits
    special_characters = string.punctuation
    all_characters = uppercase_letters + lowercase_letters + digits + special_characters
    generated_password = ''.join(random.choice(all_characters) for _ in range(length))

    return generated_password

main()
```

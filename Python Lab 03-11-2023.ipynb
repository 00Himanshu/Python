Write a Python function to find the maximum and minimum elements in a list.

```python
def main():
  my_list = [2, 6, 8, 14, 3, 77, 63]

  max_val, min_val = find_max_min(my_list)

  print("Maximum of the list:", max_val)
  print("Minimum of the list:", min_val)

def find_max_min(my_list):
  max_val = my_list[0]
  min_val = my_list[0]
  for element in my_list:
    if element > max_val:
      max_val = element
    if element < min_val:
      min_val = element

  return max_val, min_val

main()
```

Given a list of stock prices over time, write a Python function to find the best time to buy and sell to maximize profit.

```python
def max_profit(prices):

  max_profit = 0
  min_price = prices[0]

  for price in prices:
    if price < min_price:
      min_price = price
    elif price - min_price > max_profit:
      max_profit = price - min_price

  return max_profit


prices = [7, 1, 5, 3, 6, 4]
max_profit = max_profit(prices)
print("The maximum profit that can be made is:", max_profit)
```

Write a Python function to find the length of the longest increasing subsequence in a list of integers.

```python
def length_of_lis(nums):
    if not nums:
        return 0

    n = len(nums)
    dp = [1] * n

    for i in range(1, n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)

nums = [10, 22, 9, 33, 21, 50, 41, 60, 80]
result = length_of_lis(nums)
print("Length of the longest increasing subsequence:", result)
```

Given an n x m chessboard, write a program to color it using two colors (e.g., black and white) such that no two adjacent cells have the same color.

```python
def main():
    n = 8
    m = 8
    chessboard = col_ches(n, m)
    for row in chessboard:
        print(row)

def col_ches(n, m):
    chessboard = []
    colors = ["black", "white"]
    for i in range(n):
        row = []
        for j in range(m):
            color = colors[(i + j) % 2]
            row.append(color)
        chessboard.append(row)
        #print(chessboard)
    return chessboard

main()
```

Write a Python program which iterates the integers from 1 to 50. For multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz".
For numbers which are multiples of both three and five print "FizzBuzz".

```python
a = 1
b = 50
lis = []

for i in range(a, b + 1):
    if i % 3 == 0 and i % 5 == 0:
        lis.append("FizzBuzz")
    elif i % 3 == 0:
        lis.append("Fizz")
    elif i % 5 == 0:
        lis.append("Buzz")
    else:
        lis.append(i)

for index, value in enumerate(lis, start=1):
    end = " " if index % 10 != 0 else "\n"
    print(value, end=end)
```

Write a Python program to calculate a dog's age in dog's years.
           Note: For the first two years, a dog year is equal to 10.5 human years. After that, each   dog year equals 4 human years.

```python
age = int(input("Enter your Dog's age: "))
dog_year = 10.5

if age <= 2:
    age_of_dog = dog_year * age
elif age > 2:
    age_of_dog = dog_year * 2 + (age - 2) * 4

print("The Age of you dog in dog's year is ", age_of_dog)
```

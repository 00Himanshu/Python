1. Create a Person class with attributes name and age. Provide a method that prints a message introducing the person.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def intro(self):
        print(f"Hey Everyone, This is {self.name} and I am {self.age} years old.")

person1 = Person("Himanshu", 21)
person1.intro()
```

2. Create a Car class with attributes make, model, and year. Then create a drive_car method and prints a driving message.

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def __str__(self):
        return f"{self.year} {self.make} {self.model}"

class Person:
    def __init__(self, name):
        self.name = name

    def drive_car(self, car):
        print(f"{self.name} has a {car}")

car1 = Car("Mahindra", "Thar", 2022)
person1 = Person("Himanshu")
person1.drive_car(car1)
```


3. Create a Circle class with a class attribute pi set to 3.14. Add an instance method calculate_area that calculates and returns the area of the circle.

```python
class Circle:
    pi = 3.14

    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        area = self.pi * (self.radius ** 2)
        return area

c = Circle(int(input("Enter the radius of circle: ")))
area = c.calculate_area()
print("The area of the circle is:", area)
```


4. Create a class  Book with attributes title, author, and publication_year. Include a display_bookdetail() method to display the book information.

```python
class Book:
    def __init__(self, title, author, year_of_publicataion):
        self.title = title
        self.author = author
        self.year = year_of_publicataion

    def display_bookdetail(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("Publication Year:", self.year)

book1 = Book("The Mirror Within: A Tapestry of Inner-Self", "Ajit Kumar", 2023)
book1.display_bookdetail()
```


5. Create a class point to represent a point in 2D space with attributes x and y. Include a distanceCalculation()method to calculate the distance between two points

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_calculation(self, other_point):
        distance = math.sqrt((self.x - other_point.x)**2 + (self.y - other_point.y)**2)
        return distance

point1 = Point(1, 2)
point2 = Point(4, 6)

distance = point1.distance_calculation(point2)
print(f"The distance between ({point1.x}, {point1.y}) and ({point2.x}, {point2.y}) is: {distance}")
```

1. Create a base class Profile with attributes username and followers. Derive two classes, FacebookProfile and TwitterProfile, from SocialMediaUser. Add methods in each derived class to post updates and display user information.

```python
class Profile:
    def __init__(self, username, followers):
        self.username = username
        self.followers = followers

class FacebookProfile(Profile):
    def post_update(self, message):
        print(f"Posting update on Facebook: {message}")

    def display_user_info(self):
        print(f"Username: {self.username}")
        print(f"Followers: {self.followers}")

class TwitterProfile(Profile):
    def post_tweet(self, tweet):
        print(f"Posting tweet: {tweet}")

    def display_user_info(self):
        print(f"Username: {self.username}")
        print(f"Followers: {self.followers}")

facebook_profile = FacebookProfile("Rashmika", 1000)
facebook_profile.post_update("This is my first Facebook post!")
facebook_profile.display_user_info()

twitter_profile = TwitterProfile("Rupali", 500)
twitter_profile.post_tweet("Just checking in from Twitter!")
twitter_profile.display_user_info()
```

2. Create a base class Product with attributes name and price. Derive two classes, Electronics and Clothing, from Product. Add methods in each derived class to display product information.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Electronics(Product):
    def display_product_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: ${self.price}")
        print("Category: Electronics")

class Clothing(Product):
    def display_product_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: ${self.price}")
        print("Category: Clothing")

electronics_product = Electronics("Laptop", 1200)
electronics_product.display_product_info()

clothing_product = Clothing("T-Shirt", 20)
clothing_product.display_product_info()
```

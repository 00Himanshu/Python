1. Create two random 3x3 matrices (A and B).
2. Perform element-wise addition, subtraction, multiplication, and division on these matrices.
3. Compute the dot product of A and the transpose of B.

```python
import numpy as np

A = np.random.rand(3, 3)
B = np.random.rand(3, 3)

print("Matrix 1:\n", A)
print("Matrix 2:\n", B)

print("Element-wise addition:\n", A + B)
print("Element-wise subtraction:\n", A - B)
print("Element-wise multiplication:\n", A * B)
print("Element-wise division:\n", A / B)

print("Dot product of A and the transpose of B:\n", np.dot(A, B.transpose()))
```

Generate a NumPy array with integers from 0 to 20.
Extract all even numbers using array slicing.
Replace all odd numbers with -1.

```python
Array1=np.arange(0,21)
even=Array1[0::2]
Array1[1::2]=[-1]

print("Even:",even)
print("Replaced all odd numbers with -1:\n", Array1)
```

Create a random 1D array with 100 elements.
Calculate the mean, median, and standard deviation of the array.
Identify the index of the maximum and minimum values in the array.

```python
import numpy as np

Array2 = np.random.rand(100)

mean_value = np.mean(Array2)
median_value = np.median(Array2)
std_deviation = np.std(Array2)

max_index = np.argmax(Array2)
min_index = np.argmin(Array2)

print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
print("Index of Maximum Value:", max_index)
print("Index of Minimum Value:", min_index)
```

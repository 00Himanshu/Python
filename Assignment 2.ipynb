1.   Travelling salesman problem

```python
import numpy as np

dist = [["A", 300, 150, 100],
        [300, "B", 230, 150],
        [150, 230, "C", 250],
        [100, 150, 250, "D"]]

cities = ["A","B","C","D"]
num_cities = len(cities)

best_route = None
min_dist = float("inf")

def generate_tours(tour, visited):
    global min_dist, best_route

    if len(tour) == num_cities:

        dist_along_route = 0
        for i in range(num_cities-1):
            curr_city = tour[i]
            next_city = tour[i+1]
            row = cities.index(curr_city)
            col = cities.index(next_city)
            dist_along_route += dist[row][col]
        dist_along_route += dist[cities.index(tour[-1])][cities.index(tour[0])]

        if dist_along_route < min_dist:
            min_dist = dist_along_route
            best_route = list(tour)
        return

    for city in cities:
        if city not in visited:
            visited.add(city)
            tour.append(city)
            generate_tours(tour,visited)
            visited.remove(city)
            tour.pop()

generate_tours([],set())
print(f"Shortest distance: {min_dist}")
print(f"Best route: {best_route}")
```

2.  Chinese Postman problem

```python
import numpy as np

dist = [["A", 300, 90, 100],
        [300, "B", 230, 150],
        [90, 230, "C", 250],
        [100, 150, 250, "D"]]

cities = ["A", "B", "C", "D"]
num_cities = len(cities)

best_route = None
min_dist = float("inf")

def generate_tours(tour, visited):
    global min_dist, best_route

    if len(tour) == num_cities:
        dist_along_route = 0
        for i in range(num_cities - 1):
            curr_city = tour[i]
            next_city = tour[i + 1]
            row = cities.index(curr_city)
            col = cities.index(next_city)
            dist_along_route += dist[row][col]
        dist_along_route += dist[cities.index(tour[-1])][cities.index(tour[0])]

        if dist_along_route < min_dist:
            min_dist = dist_along_route
            best_route = list(tour)
        return

    for i in range(num_cities):
        for j in range(num_cities):
            if i != j:
                city1 = cities[i]
                city2 = cities[j]
                if (city1 not in visited) or (city2 not in visited):
                    visited.add(city1)
                    visited.add(city2)
                    tour.append((city1, city2))
                    generate_tours(tour, visited)
                    visited.remove(city1)
                    visited.remove(city2)
                    tour.pop()

generate_tours([], set())
print(f"Shortest distance: {min_dist}")
print(f"Best route: {best_route}")
```

3.   Towers of Hanoi Problem

```python
def tower_of_hanoi(num_disks, from_rod, to_rod, aux_rod):
    if num_disks == 1:
        print(f"Move disk 1 from rod {from_rod} to rod {to_rod}")
        return

    tower_of_hanoi(num_disks - 1, from_rod, aux_rod, to_rod)
    print(f"Move disk {num_disks} from rod {from_rod} to rod {to_rod}")
    tower_of_hanoi(num_disks - 1, aux_rod, to_rod, from_rod)


num_disks = 3
from_rod = 'A'
to_rod = 'C'
aux_rod = 'B'

tower_of_hanoi(num_disks, from_rod, to_rod, aux_rod)
```

4.   Missionaries and cannibals problem

```python
Max_capacity = 2

def solution(Left_side, Right_side):
    Right_M = 0
    Right_C = 0
    for p in Right_side:
        if p == "M":
            Right_M += 1
        if p == "C":
            Right_C += 1

    while Right_C > 0:
        if Right_C >= Max_capacity:
            num_to_move = Max_capacity
        else:
            num_to_move = Right_C

        if Right_M >= num_to_move and Right_M >= Right_C:
            m_to_move = num_to_move
        else:
            m_to_move = Right_M
        make_trip(num_to_move, m_to_move)

def make_trip(C, M):
    global Left_side, Right_side

    for _ in range(M):
        if "M" in Right_side:
            Right_side.remove("M")
        Left_side.append("M")
    for _ in range(C):
        if "C" in Right_side:
            Right_side.remove("C")
        Left_side.append("C")

Right_side = ["M", "M", "M", "X", "X", "X"]
Left_side = []
solution(Left_side, Right_side)
print("Left side:", Left_side)
print("Right side:", Right_side)
```

Write a Python program to calculate the number of days between two dates.

```python
from datetime import datetime
def main():
  date_str1 = input("Enter the First date in DD-MM-YYYY: ")
  date_str2 = input("Enter the Second date in DD-MM-YYYY: ")

  result = days_between_dates(date_str1, date_str2)
  print(f"Number of days between {date_str1} and {date_str2}: {result} days")
def days_between_dates(date1, date2):

    date_format = "%d-%m-%Y"
    datetime1 = datetime.strptime(date1, date_format)
    datetime2 = datetime.strptime(date2, date_format)
    print(datetime1, datetime2)

    delta = datetime2 - datetime1

    return delta.days
main()
```

Write a Python program to find out how many days have been left in this year.

```python
from datetime import datetime

def main():
    date_str = input("Enter the date in DD-MM-YYYY format: ")
    remaining_days = days_left_in_year(date_str)
    print(f"Number of days left in the year from {date_str}: {remaining_days} days")

def days_left_in_year(date_str):
    date_format = "%d-%m-%Y"
    input_date = datetime.strptime(date_str, date_format)

    end_of_year = datetime(input_date.year, 12, 31)


    delta = end_of_year - input_date

    return delta.days

main()
```

```python
def pyramid(a):
  for i in range(5,0,-1):
    print((a + "    ")*i + "\n")

pyramid(input("Type one character to make pyramid: "))
```

```python
def print_isosceles_triangle(height):
    for i in range(1, height + 1):
        spaces = " " * (height - i)
        stars = "*" * (2 * i)
        print(spaces + stars)

# Input the height of the triangle
triangle_height = int(input("Enter the height of the isosceles triangle: "))

# Call the function to print the triangle
print_isosceles_triangle(triangle_height)
```

4    4    4    4

1

```python
def number_pyramid(n):
  if n>=0:
    for i in range (n,0,-1):
      print(((str(i))+"     ")*i+"\n")
  else: print("Enter number greater then 0 ")

number_pyramid(int(input("Enter a number: ")))
```

```python
def number_pyramid(n):
  if n>=0:

    for i in range (n,0,-1):
      for j in range(i):
        print(i, end=" ")
      print()

  else: print("Enter number greater then 0 ")

number_pyramid(int(input("Enter a number: ")))
```

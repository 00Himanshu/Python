3. Write a program for creating the functionality of the restaurant class. It has the below user-defined functions

setmenu() function for inputting the item and its cost

getmenu() function for getting the menu item and cost

getorder() function for selecting the item from the menu

getcost() function for getting the total cost of the selected item

```python
class Restaurant:
    def __init__(self):
        self.menu = {}

    def setmenu(self):
        print("Enter the menu items and their prices:")
        for i in range(1, 7):
            item = input(f"Item {i}: ")
            price = float(input(f"Price for {item}: "))
            self.menu[item] = price

    def getmenu(self):
        print("Menu:")
        for item, price in self.menu.items():
            print(f"{item}: {price:.2f}")

    def getorder(self):
        print("Select items from the menu:")
        order = input("Enter item (type 'done' to finish): \n")
        selected_items = []

        while order.lower() != 'done':
            if order in self.menu:
                selected_items.append(order)
            else:
                print(f"{order} is not on the menu. Please select a valid item.")

            order = input("Enter another item (type 'done' to finish): \n")

        return selected_items

    def getcost(self, selected_items):
        total_cost = sum(self.menu[item] for item in selected_items)
        return total_cost




my_restaurant = Restaurant()

my_restaurant.setmenu()

my_restaurant.getmenu()

order_items = my_restaurant.getorder()

total_cost = my_restaurant.getcost(order_items)
print(f"Total Cost: {total_cost:.2f}")
```

```python
class ElectricityBill:
    def __init__(self):
        self.ebillno = 0
        self.ename = None
        self.units = 0
        self.subcharge = 0
        self.place = None
        self.amount = 0.0

    def ksebbillinfo(self):
        self.ebillno = int(input("Enter Electricity Bill Number: "))
        self.ename = input("Enter Customer Name: ")
        self.units = int(input("Enter Units Consumed: "))
        self.subcharge = int(input("Enter Subcharge: "))
        self.place = input("Enter Place (village/corporation/metrocity): ")

    def ksebbillcalculations(self):
        if self.place == "village" and 110 <= self.units < 135:
            self.amount = self.units + (self.units * 18 / 100) + self.subcharge
        elif self.place == "corporation" and 135 <= self.units < 150:
            self.amount = self.units + (self.units * 25 / 100) + self.subcharge
        elif self.place == "metrocity" and self.units >= 150:
            self.amount = self.units + (self.units * 40 / 100) + self.subcharge

    def ksebelectricitybill(self):
        print("\nElectricity Bill Details:")
        print("EBill Number:", self.ebillno)
        print("Customer Name:", self.ename)
        print("Units Consumed:", self.units)
        print("Subcharge:", self.subcharge)
        print("Place:", self.place)
        print("Amount:", self.amount)



electricity_bill_instance = ElectricityBill()

electricity_bill_instance.ksebbillinfo()

electricity_bill_instance.ksebbillcalculations()

electricity_bill_instance.ksebelectricitybill()
```

5. Write a program to build a simple Student Management System using Python, which can perform the following operations:

Accept(): This method takes details from the user like name, roll number, and marks for two different subjects.

Display(): This method displays the details of every student.

Search(): This method will ask the user for the roll number and then search according to the roll number

Delete(): This method deletes the record of a particular student with a matching roll number.

Update(): This method updates the roll number of the student. This method will ask for the old roll number and the new roll number. It will replace the old roll number with the new roll number

```python
class Student:
    def __init__(self, name, roll_number, subject1_marks, subject2_marks):
        self.name = name
        self.roll_number = roll_number
        self.subject1_marks = subject1_marks
        self.subject2_marks = subject2_marks

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Subject 1 Marks: {self.subject1_marks}")
        print(f"Subject 2 Marks: {self.subject2_marks}")
        print("")

class StudentManagementSystem:
    def __init__(self):
        self.students = []

    def accept(self):
        name = input("Enter student name: ")
        roll_number = int(input("Enter roll number: "))
        subject1_marks = float(input("Enter marks for Subject 1: "))
        subject2_marks = float(input("Enter marks for Subject 2: "))

        student = Student(name, roll_number, subject1_marks, subject2_marks)
        self.students.append(student)
        print("Student details added successfully!\n")

    def display(self):
        if not self.students:
            print("No student records found.\n")
            return

        print("Student Details:")
        for student in self.students:
            student.display_details()

    def search(self, roll_number):
        for student in self.students:
            if student.roll_number == roll_number:
                student.display_details()
                return

        print(f"No student found with Roll Number {roll_number}\n")

    def delete(self, roll_number):
        for student in self.students:
            if student.roll_number == roll_number:
                self.students.remove(student)
                print(f"Student with Roll Number {roll_number} deleted successfully!\n")
                return

        print(f"No student found with Roll Number {roll_number}\n")

    def update(self, old_roll_number, new_roll_number):
        for student in self.students:
            if student.roll_number == old_roll_number:
                student.roll_number = new_roll_number
                print(f"Roll Number updated successfully!\n")
                return

        print(f"No student found with Roll Number {old_roll_number}\n")

student_system = StudentManagementSystem()

while True:
    print("Student Management System Menu:")
    print("1. Accept Student Details")
    print("2. Display Student Details")
    print("3. Search Student by Roll Number")
    print("4. Delete Student by Roll Number")
    print("5. Update Student Roll Number")
    print("6. Exit")

    choice = int(input("Enter your choice (1-6): "))

    if choice == 1:
        student_system.accept()
    elif choice == 2:
        student_system.display()
    elif choice == 3:
        roll_number = int(input("Enter Roll Number to search: "))
        student_system.search(roll_number)
    elif choice == 4:
        roll_number = int(input("Enter Roll Number to delete: "))
        student_system.delete(roll_number)
    elif choice == 5:
        old_roll_number = int(input("Enter current Roll Number: "))
        new_roll_number = int(input("Enter new Roll Number: "))
        student_system.update(old_roll_number, new_roll_number)
    elif choice == 6:
        print("Exiting Student Management System. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.\n")
```

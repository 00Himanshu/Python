```python
class Animal:
  def __init__(self, species):
    self.species = species

  def display_species(self):
    print(f"Species: {self.species}")

class Soundmixin:
  def __init__(self, sound):
    self.sound= sound

  def make_sound(self):
    print(f"Making the sound: {self.sound}")

class Pet(Animal, Soundmixin):
  def __init__(self, species, sound, name):
    Animal.__init__(self, species)
    Soundmixin.__init__(self, sound)
    self.name= name

  def display_info(self):
    self.display_species()
    self.make_sound()
    print(f"Name: {self.name}")

pet1 = Pet("Dog", "Woof", "Buddy")
pet1.display_info()
```

```python
class Animal:
  def __init__(self, species):
    self.species = species

  def display_species(self):
    print(f"Species: {self.species}")

class Dog(Animal):
  def __init__ (self, species, breed):
    super().__init__(species)
    self.breed = breed

  def display_info(self):
    self.display_species()
    print(f"Breed: {self.breed}")

class Cat(Animal):
  def __init__ (self, species, color):
    super().__init__(species)
    self.color = color

  def display_info(self):
    self.display_species()
    print(f"Color: {self.color}")

dog1=Dog("Canine", "Labrador")
dog1.display_info()

cat1=Cat("Feline", "Siamese")
cat1.display_info()
```

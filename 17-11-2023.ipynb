3.Write a program that counts the number of vowels in a given string.

```python
def main():
  str1=input("Enter the massage: ")
  print(f"Number of vowels persent in this massage is {CountVowels(str1)}")
def CountVowels(msg):
  count=0
  vowels=["a", "e", "i", "o", "u"]
  for s in msg:
    if s.lower() in vowels:
      count+=1
  return count
main()
```

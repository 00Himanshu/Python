1. WAP in python to implement calculator using function overloading

```python
def calculator(ops, *args):
  answer = args[0]

  if ops == "add":
      for x in args[1:]:
          answer += x

  elif ops == "sub":
      for x in args[1:]:
          answer -=  x

  elif ops == "divide":
      for x in args[1:]:
          answer /=  x

  elif ops == "multi":
      for x in args[1:]:
          answer *=  x

  print(answer)

calculator("add", 5, 6, 7, 7)
calculator("sub", 5, 6, 7, 8)
calculator("divide", 6, 6)
calculator("multi", 5, 6, 7)
```

2.WAP in python to calculate salary of an employee using function overloading.

```python
def calculate_salary(**kwargs):
    name = kwargs.get("name")
    basic_salary = kwargs.get("basic_salary")
    bonus = kwargs.get("bonus", 0)
    allowance = kwargs.get("allowance", 0)
    absents = kwargs.get("absents", 0)

    if name is None or basic_salary is None:
        return "Invalid input. Name and basic_salary are required."

    total_salary = basic_salary + bonus + allowance - (absents * (basic_salary//30))
    return f"{name}'s total salary: {total_salary}"

salary_without_bonus = calculate_salary(name="Himanshu", basic_salary=50000)
salary_with_bonus = calculate_salary(name="Sivakami", basic_salary=50000, bonus=10000)
salary_with_allowance = calculate_salary(name="Bob", basic_salary=50000, allowance=5000)
salary_with_absents = calculate_salary(name="Ajit", basic_salary=50000, absents=2)

print(salary_without_bonus)
print(salary_with_bonus)
print(salary_with_allowance)
print(salary_with_absents)
```

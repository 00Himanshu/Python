1. Create a base class Profile with attributes username and followers. Derive two classes, FacebookProfile and TwitterProfile, from SocialMediaUser. Add methods in each derived class to post updates and display user information.

```python
class Profile:
    def __init__(self, username, followers):
        self.username = username
        self.followers = followers

class FacebookProfile(Profile):
    def post_update(self, message):
        print(f"Posting update on Facebook: {message}")

    def display_user_info(self):
        print(f"Username: {self.username}")
        print(f"Followers: {self.followers}")

class TwitterProfile(Profile):
    def post_tweet(self, tweet):
        print(f"Posting tweet: {tweet}")

    def display_user_info(self):
        print(f"Username: {self.username}")
        print(f"Followers: {self.followers}")

facebook_profile = FacebookProfile("Rashmika", 1000)
facebook_profile.post_update("This is my first Facebook post!")
facebook_profile.display_user_info()

twitter_profile = TwitterProfile("Rupali", 500)
twitter_profile.post_tweet("Just checking in from Twitter!")
twitter_profile.display_user_info()
```

2. Create a base class Product with attributes name and price. Derive two classes, Electronics and Clothing, from Product. Add methods in each derived class to display product information.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Electronics(Product):
    def display_product_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: ${self.price}")
        print("Category: Electronics")

class Clothing(Product):
    def display_product_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: ${self.price}")
        print("Category: Clothing")

electronics_product = Electronics("Laptop", 1200)
electronics_product.display_product_info()

clothing_product = Clothing("T-Shirt", 20)
clothing_product.display_product_info()
```

```python
from multipledispatch import dispatch

@dispatch (int, int)
def product(first, second):
  result = first*second
  print(result)

@dispatch (int, int, int)
def product(first, second, third):
  result = first * second * third
  print(result)

@dispatch (float, float, float)
def product (first, second, third):
  result= first * second * third
  print(result)

product(2, 3)

product(2,3,2)

product(2.2, 3.4, 2.3)
```

```python
class Point:
  def __init__(self, x=0, y=0):
    self.x=x
    self.y=y

  def __str__(self):
    return "({0}, {1})".format(self.y,self.x)

  def __add__(self, other):
    x=self.x + other.x
    y=self.y + other.y
    return Point(x,y)

p1= Point(1,3)
p2= Point(2,4)
p3= Point(5,6)
print(p1+p2)
print(p1+p2+p3)
```

WAP in python to calculate salary of an employee using function overloading

```python
def calculate_salary(name=None, basic_salary=None, bonus=0, allowance=0, absents=0):
    if name is None or basic_salary is None:
        return "Invalid input. Name and basic_salary are required."

    total_salary = basic_salary + bonus + allowance - (absents * 1000)
    return f"{name}'s total salary: {total_salary}"

salary_without_bonus = calculate_salary(name="Himanshu", basic_salary=50000)
salary_with_bonus = calculate_salary(name="Sivakami", basic_salary=50000, bonus=10000)
salary_with_allowance = calculate_salary(name="Bob", basic_salary=50000, allowance=5000)
salary_with_absents = calculate_salary(name="Ajit", basic_salary=50000, absents=2)


print(salary_without_bonus)
print(salary_with_bonus)
print(salary_with_allowance)
print(salary_with_absents)
```

```python
class Employee:

  def salary(self, basic_salary, bonus=0):
    self.basic_salary= basic_salary
    self.bonus = bonus
    total_salary = self.basic_salary + self.bonus
    return total_salary

  def salary_allowance(self, basic_salary, allowance):
    self.allowance = allowance
    total_salary = self.basic_salary + self.allowance
    return total_salary

employee=Employee()
s_w_o_b= employee.salary(50000)
s_w_b= employee.salary(50000, 10000)
s_w_a= employee.salary_allowance(50000, 5000)
print("Salary without bonus: ", s_w_o_b)
print("Salary with bonus:", s_w_b)
print("Salary with allownce:", s_w_a)
```

```python
class Person:
  def __init__(self, name, age):
    self.name=name
    self.age=age
  def display_info(self):
    print(f'Name: {self.name}')
    print(f'Age: {self.age}')



class Student(Person):
  def __init__(self, name, age, student_id):
    super().__init__(name, age)
    self.student_id= student_id

  def display_student(self):
    print(f'Student Name: {self.name}')
    print(f'Student Age: {self.age}')
    print(f'Student Id: {self.student_id}')

class Teacher(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def display_teacher(self):
        print(f'Teacher Name: {self.name}')
        print(f'Teacher Age: {self.age}')
        print(f'Employee Id: {self.employee_id}')

student1 = Student("Himanshu", 21, 231024)
teacher1 = Teacher("Judy", 32, 210223)

student1.display_student()
print(" ")
teacher1.display_teacher()
```

4. You are developing a zoo simulation program. Each animal in the zoo is classified as either a mammal or a bird. All animals share common attributes, and specific attributes distinguish mammals (such as fur color) and birds (such as feather color). Your program should reflect this class hierarchy and provide flexibility for adding new types of animals in the future.

```python
class Zoo:
    def __init__(self, name, kingdom, phylum):
        self.name = name
        self.kingdom = kingdom
        self.phylum = phylum

class Mammal(Zoo):
    def __init__(self, name, kingdom, phylum, species, breed, sound, fur_color):
        super().__init__(name, kingdom, phylum)
        self.species = species
        self.breed = breed
        self.sound = sound
        self.fur_color = fur_color

    def print_info(self):
        print(f"Species: {self.species}, Breed: {self.breed}, Sound: {self.sound}, Fur Color: {self.fur_color}")

class Bird(Zoo):
    def __init__(self, name, kingdom, phylum, species, breed, feather_color, sound):
        super().__init__(name, kingdom, phylum)
        self.species = species
        self.breed = breed
        self.feather_color = feather_color
        self.sound = sound

    def print_info(self):
        print(f"Species: {self.species}, Breed: {self.breed}, Feather Color: {self.feather_color}, Sound: {self.sound}")


lion = Mammal("Lion", "Animalia", "Chordata", "Panthera leo", "African Lion", "Roar", "Golden")
eagle = Bird("Eagle", "Animalia", "Chordata", "Aquila", "Golden Eagle", "Brown", "Screech")


lion.print_info()
eagle.print_info()
```

5.You are designing a system for a library that involves multiple levels of book categorization. The base class is Book, and you have derived classes like FictionBook and NonFictionBook. Now, there's a new requirement to introduce a more specific category for ScienceFictionBook. How would you structure the class hierarchy to accommodate this new category while ensuring code readability and maintainability?

```python
class Book:
    def __init__(self, title, author, genre):
        self.title = title
        self.author = author
        self.genre = genre

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}, Genre: {self.genre}")


class FictionBook(Book):
    def __init__(self, title, author, genre, fiction_type):
        super().__init__(title, author, genre)
        self.fiction_type = fiction_type

    def display_info(self):
        super().display_info()
        print(f"Fiction Type: {self.fiction_type}")


class NonFictionBook(Book):
    def __init__(self, title, author, genre, topic):
        super().__init__(title, author, genre)
        self.topic = topic

    def display_info(self):
        super().display_info()
        print(f"Topic: {self.topic}")


class ScienceFictionBook(FictionBook):
    def __init__(self, title, author, fiction_type, subgenre):
        super().__init__(title, author, "Science Fiction", fiction_type)
        self.subgenre = subgenre

    def display_info(self):
        super().display_info()
        print(f"Subgenre: {self.subgenre}")


fiction_book = FictionBook("The Great Gatsby", "F. Scott Fitzgerald", "Classic", "Novel")
non_fiction_book = NonFictionBook("Sapiens", "Yuval Noah Harari", "History", "Human Evolution")
science_fiction_book = ScienceFictionBook("Dune", "Frank Herbert", "Space Opera", "Epic")

print("Fiction Book Information:")
fiction_book.display_info()

print("\nNon-Fiction Book Information:")
non_fiction_book.display_info()

print("\nScience Fiction Book Information:")
science_fiction_book.display_info()
```
